```python
import pandas as pd
```

```python
fifa_data = pd.read_csv("fifa21.csv", low_memory=False)
fifa_data.head()
```

```python
# Display basic information about the dataset
print("Original Data:")
print(fifa_data.info())
```

```python
# Handling missing values
# For simplicity, let's drop rows with missing values
#fifa_data = fifa_data.dropna()
```

```python
# Display the column names
print("Column Names:")
print(fifa_data.columns)
```

```python
# Drop the columns 'photoUrl' and 'playerUrl'
columns_to_drop = ['photoUrl', 'playerUrl']
fifa_data = fifa_data.drop(columns=columns_to_drop)
```

```python
# Convert 'Value' and 'Wage' columns to numeric
fifa_data['Value'] = fifa_data['Value'].replace('[\€KM]', '', regex=True).apply(lambda x: 1e3 if 'K' in x else 1e6).astype(float) * fifa_data['Value'].replace('[\€KM]', '', regex=True).astype(float)
fifa_data['Wage'] = fifa_data['Wage'].replace('[\€KM]', '', regex=True).apply(lambda x: 1e3 if 'K' in x else 1e6).astype(float) * fifa_data['Wage'].replace('[\€KM]', '', regex=True).astype(float)
```

```python
# Convert 'Release Clause' column to numeric
fifa_data['Release Clause'] = fifa_data['Release Clause'].replace('[\€KM]', '', regex=True).apply(lambda x: 1e3 if 'K' in x else 1e6).astype(float) * fifa_data['Release Clause'].replace('[\€KM]', '', regex=True).astype(float)
```

```python
fifa_data["Contract"].head()
```

```python
# Split the 'Contract' column into two columns using the '~' delimiter
fifa_data[['Contract Start', 'Contract End']] = fifa_data['Contract'].str.split('~', expand=True)

# Drop the original 'Contract' column
fifa_data = fifa_data.drop(columns=['Contract'])
```

```python
fifa_data[['Contract Start', 'Contract End']].head()
```

```python
# Fill empty cells in the 'Loan Date' column with 'NA'
fifa_data['Loan Date End'] = fifa_data['Loan Date End'].fillna('NA')
fifa_data['Loan Date End'].head()
```

```python
fifa_data["Height"].head()
```

```python
# Remove "cm" from the 'Height' column
fifa_data['Height'] = fifa_data['Height'].str.replace('cm', '')

# Rename the 'Height' column to 'Height (cm)'
fifa_data = fifa_data.rename(columns={'Height': 'Height(cm)'})
fifa_data['Height(cm)'].head()
```

```python
# Identify non-numeric values in "Height(cm)" column
non_numeric_heights = pd.to_numeric(fifa_data["Height(cm)"], errors='coerce').isna()
fifa_data[non_numeric_heights]['Height(cm)'].head(3)
```

```python
# Convert "Height(cm)" column to centimeters
def convert_height(height):
    try:
        # Extract feet and inches
        feet, inches = map(int, height.replace("\"", "").split("'"))
        # Convert to centimeters
        total_centimeters = feet * 30.48 + inches * 2.54
        return total_centimeters
    except (ValueError, TypeError):
        return None  # Return None for cases where conversion fails
```

```python
fifa_data["Height(cm)"] = fifa_data["Height(cm)"].apply(convert_height)
```

```python
fifa_data['Weight'].head()
```

```python
# Remove "kg" from the 'Weight' column
fifa_data['Weight'] = fifa_data['Weight'].str.replace('kg', '')
```

```python
# Rename the 'Weight' column to 'Weight(kg)'
fifa_data = fifa_data.rename(columns={'Weight': 'Weight(kg)'})
fifa_data['Weight(kg)'].head()
```

```python
# Remove "lbs" from "Weight(kg)" column, convert to numeric, and convert pounds to kilograms
fifa_data["Weight(kg)"] = pd.to_numeric(fifa_data["Weight(kg)"].str.replace("lbs", ""), errors='coerce')
fifa_data["Weight(kg)"] = fifa_data["Weight(kg)"] * 0.453592  # 1 pound = 0.453592 kilograms
```

```python
fifa_data["W/F"].head()
```

```python
# Extract only the numeric part from the 'W/F' column
fifa_data['W/F'] = fifa_data['W/F'].str.extract('(\d+)')
fifa_data['W/F'].head()
```

```python
# Rename the 'W/F' column to 'W/F(stars)'
fifa_data = fifa_data.rename(columns={'W/F': 'W/F(stars)'})
fifa_data['W/F(stars)'].head()
```

```python
fifa_data['SM'].head()
```

```python
# Extract only the numeric part from the 'SM' column
fifa_data['SM'] = fifa_data['SM'].str.extract('(\d+)')

# Rename the 'W/F' column to 'W/F(stars)'
fifa_data = fifa_data.rename(columns={'SM': 'SM(stars)'})
fifa_data['SM(stars)'].head()
```

```python
# Extract only the numeric part from the 'IR' column
fifa_data['IR'] = fifa_data['IR'].str.extract('(\d+)')

# Rename the 'IR' column to 'IR(stars)'
fifa_data = fifa_data.rename(columns={'IR': 'IR(stars)'})
fifa_data['IR(stars)'].head()
```

```python
fifa_data['Hits'].head(10)
```

```python
# Define a function to process the 'Hits' column
def process_hits(hits):
    if isinstance(hits, float):
        return hits
    elif 'K' in hits:
        return float(hits.replace('K', '')) * 1000
    else:
        return float(hits)

# Apply the function to the 'Hits' column
fifa_data['Hits'] = fifa_data['Hits'].apply(process_hits)

fifa_data['Hits'].head(10)
```

```python
# Display the cleaned dataset
print("\nCleaned Data:")
print(fifa_data.info())
```

```python
fifa_data.head()
```

```python
# Remove newline characters from the 'Club' column
fifa_data['Club'] = fifa_data['Club'].str.replace('\n', '')
fifa_data.head()
```

```python
# Split the 'Positions' column into multiple columns based on commas
positions_split = fifa_data['Positions'].str.split(',', expand=True)
```

```python
# Display the DataFrame with the new columns
fifa_data = pd.concat([fifa_data, positions_split], axis=1)
fifa_data
```

```python
# Clean each new position column (e.g., removing leading/trailing whitespaces)
for col in positions_split.columns:
    fifa_data[col] = fifa_data[col].str.strip()
```

```python
# Drop the original 'Positions' column and unnecessary intermediate columns
fifa_data = fifa_data.drop(['Positions'], axis=1)
```

```python
# Rename the specified columns
column_mapping = {0: 'position 1', 1: 'position 2', 2: 'position 3'}
fifa_data = fifa_data.rename(columns=column_mapping)
```
